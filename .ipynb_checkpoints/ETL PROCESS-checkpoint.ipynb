```python
def read_columns(c):
 
    # initialize an empty string
    str1 = ""
 
    # traverse in the string
    for i,ele in enumerate(c):
        if i != len(c)-1:
            str1 += f"{ele},"
        else:
            str1 += ele
    # return string
    return str1
```

```python
from sqlalchemy import create_engine, MetaData

# Create an engine to connect to the database
engine = create_engine(f"postgresql://{user}:{password}@localhost:5432/{database}")

# Create a metadata object and reflect the schema
metadata = MetaData()
metadata.reflect(bind=engine)

# Specify the schema name
schema_name = 'raw_staging'

# Generate the SQL statements to create the schema in the database
table_sql = []
for table in metadata.sorted_tables:
    table_sql.append(table.create(bind=engine, checkfirst=False, schema=schema_name))

schema_sql = "\n".join(table_sql)

# Write the SQL statements to a file
with open(f'{schema_name}.sql', 'w') as f:
    f.write(schema_sql)
```
