```python
def db_init_connection(host,port, database, user, password):
    """
    Creates a database connection and returns the connection object.
    """
    conn = psycopg2.connect(
        host=host,
        port=port,
        database=database,
        user=user,
        password=password
    )
    return conn
```

```python
import csv
import os

def read_csv_files(folder_path):
    file_paths = []
    print(len(file_paths))

    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_paths.append(os.path.join(folder_path, filename))

    for file_path in file_paths:
        with open(file_path, 'r') as f:
            reader = csv.reader(f)
            next(reader)  # skip the header row
            for row in reader:
                print(row)
```

```python
import numpy as np

def get_all_column_names(folder_path):
    file_paths = []
    column_names = set()

    # Find all CSV files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_paths.append(os.path.join(folder_path, filename))

    # Read all CSV files and get the column names
    for file_path in file_paths:
        with open(file_path, 'r') as f:
            reader = csv.reader(f)
            header = next(reader)
            # Add the column names to the set
            column_names.update(set(header))

    # Convert the set of column names to a sorted numpy array
    return np.array(sorted(list(column_names)))
```

```python
get_all_column_names("data")
```

```python
import os
import pandas as pd

def extract_data_from_folder(folder_path):
    # Get a list of all CSV files in the folder
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

    # Read each CSV file into a Pandas DataFrame and concatenate them into a single DataFrame
    dfs = []
    for csv_file in csv_files:
        csv_path = os.path.join(folder_path, csv_file)
        df = pd.read_csv(csv_path)
        dfs.append(df)
    df = pd.concat(dfs, ignore_index=True)

    return df
```

```python
extract_data_from_folder("data")
```
